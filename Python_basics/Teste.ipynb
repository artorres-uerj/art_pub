```
x <- seq(1,50,1)
y <- rnorm(length(x),0,2)
plot(x,y)
```

```
dados <- read.csv(file = "https://raw.githubusercontent.com/artorres-uerj/art_pub/master/dados_teste.csv", sep = ",", dec = ".", header = TRUE)
str(dados)
```

```
plot(dados$x, dados$y)
fit <- lm(y ~ x, data = dados)
summary(fit)
dados$y_pred <- predict(fit)
lines(dados$x, dados$y_pred, col = "blue", lty = 2 )
```
